```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, f1_score, roc_auc_score, recall_score
import matplotlib.pyplot as plt
import json
from keras.models import load_model
from keras.layers import Input, Concatenate, Dense
from keras.models import Model
from keras.optimizers import Adam
```

```python
directory_path = r""
results_df = pd.DataFrame(columns=['File', 'Accuracy', 'Precision', 'F1 Score', 'AUC Score', 'Recall'])
label_df = pd.DataFrame(columns=['True Labels', 'Predicted'])

timestamps_df = pd.DataFrame(columns=['Timestamp (quake)', 'True Labels'])

# List all CSV files in the directory
file_paths = [os.path.join(directory_path, file) for file in os.listdir(directory_path) if file.endswith(".csv")]
results_list = []
abs_val = []
data_list = []
timestamps_list = []

true_labels_list = []
predicted_labels_list = []

for file_path in file_paths:
    df3 = pd.DataFrame()
    new_dataset = pd.read_csv(file_path, delimiter=';')
    df3 = new_dataset
    true_labels = pd.DataFrame(columns=['Class'])
    df3.drop('Data attribute 1', axis=1, inplace=True)

    if 'Data attribute 2' in df3.columns and 'Data attribute 3' in df3.columns:
        df3.drop(['Data attribute 2', 'Data attribute 3'], axis=1, inplace=True)

    true_labels['Class'] = df3['Labeling 1']
    true_labels['Class'] = true_labels['Class'].replace({'Rest': 0, 'quake': 1}).astype(int)
    df3.drop('Labeling 1', axis=1, inplace=True)

    df3.rename(columns={'Data attribute 4': 'Value'}, inplace=True)
    df3['Value'] = df3['Value'].str.replace(',', '.').astype(float)

    min_val = df3['Value'].min()
    max_val = df3['Value'].max()

    df3['NORMALIZED'] = ((df3['Value'] - min_val) / (max_val - min_val)) * 2 - 1
    df3.drop('Value', axis=1, inplace=True)

    # Assuming you have already defined 'model' for prediction
    model_car = load_model(r"E:\Leo\Documents\SHARING\Iomt\quakes_fold\models_from_training\final_model_car.h5")
    model_quake = load_model(r"E:\Leo\Documents\SHARING\Iomt\quakes_fold\models_from_training\final_model_quake.h5")
    input_layer = Input(shape=(1, 1))
    output1 = model_car(input_layer)
    output2 = model_quake(input_layer)
    concatenated = Concatenate()([output1, output2])
    concatenated_model = Model(inputs=input_layer, outputs=concatenated)
    output_layer = Dense(1, activation='sigmoid')(concatenated)
    ensemble_model = Model(inputs=input_layer, outputs=output_layer)
    ensemble_model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
    predictions = concatenated_model.predict(df3)

    # Extract the predicted classes (assuming it's a classification task)
    predicted_classes = np.argmax(predictions, axis=1)
    # Add the predicted classes as a new column in the DataFrame
    df3['predicted_class'] = predicted_classes
    df3.to_csv(r'E:\Leo\Documents\SHARING\Iomt\FinalDatasetNoClass.csv')
    label_df = pd.concat([label_df, pd.DataFrame({'True Labels': true_labels['Class'].tolist(), 'Predicted': predicted_classes.tolist()})], ignore_index=True)

    import matplotlib.pyplot as plt
    df_output = pd.DataFrame
    df_output = pd.read_csv(r'')
    print('File:', file_path)
    start = df3.query('predicted_class==1')['NORMALIZED'].index[0]
    end = df3.query('predicted_class==1')['NORMALIZED'].index[-1]
    start_true = true_labels.query('Class==1').index[0]
    end_true = true_labels.query('Class==1').index[-1]
    
    index_in_pred_labels = df3.query('predicted_class == 1').index[0]
    actual_quake_value = true_labels.iloc[index_in_pred_labels]['Class']

    plt.figure(figsize=(10, 6))
    plt.plot(df3.NORMALIZED)
    plt.plot(df3.NORMALIZED[start:end], color='r')
    plt.axvline(x=start_true, color='g', linestyle='--', label='Start (True)')
    plt.axvline(x=end_true, color='r', linestyle='--', label='End (True)')
    plt.xlabel('Index')
    plt.ylabel('Normalized Values')
    plt.legend()
    plt.show()

    # Store the timestamp of the first "quake" found and its corresponding true label
    if start_true >= 0:
        timestamps_list.append({'Timestamp (quake)': index_in_pred_labels, 'True Labels': actual_quake_value})

    # ... rest of your code ...

# After processing all files, you can access the timestamps_df DataFrame to see the timestamp of the first "quake" found and its corresponding true label.
print(timestamps_df)
```
